# TASK #4: CREATE TESTING AND TRAINING DATASET & PERFORM DATA CLEANING

```python
x_cat = emp.select_dtypes(include=['object'])
x_cat
```

```python
for i in x_cat.columns:
    print(f'{i} - {emp[i].unique()}')
    
```

```python
# encoding categorical variables
emp = pd.get_dummies(emp,columns=['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole',
       'MaritalStatus'], drop_first=True).astype(int)
```

```python
emp.info()
```

```python
# note that we dropped the target 'Atrittion'
x = emp.drop('Attrition',axis=1)
y = emp['Attrition']
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
for i in x.columns:
    #emp[i] = sc.fit_transform(emp[[i]])
    x[i] = sc.fit_transform(x[[i]])
    
```

# Data Imbalance -


```python
# METHOD 1
from imblearn.over_sampling import ADASYN
ada = ADASYN(random_state=42)
X_train_resampled, y_train_resampled = ada.fit_resample(x, y)
```

```python
# METHOD 2
# import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_resample, y_resample = smote.fit_resample(x,y)
```

```python
y_resample.value_counts()
```

```python
X_train_resampled.shape
```

```python
y_train_resampled.shape
```

```python
y_train_resampled.value_counts()
```

# TASK 5: TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_train_resampled, y_train_resampled,test_size=0.3,random_state=21)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

# TASK 7: TRAIN AND EVALUATE A DEEP LEARNING MODEL

```python
# import tensorflow as tf
# from tensorflow import keras
# from keras.layers import Dense
# from keras.layers import Dropout
# from tensorflow.keras.models import Sequential
# # Initialize the Sequential Model
# model = Sequential()
# model.add(Dense(128, input_dim=2, activation = 'relu'))
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=300,activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.01),input_shape=(44,))) # input_dim=44 # no of independent columns
model.add(tf.keras.layers.Dense(units=200,activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam',loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
hist = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=44)
# batch_size = number of training examples utilized in one iteration, no of rows.
# bgd = entire, sgd = 1 , mini gd - 1<mini<bgd 
```

```python
# Testing Set Performance without smote
plt.plot(hist.history['loss'],label='train')
plt.plot(hist.history['val_loss'],label='test')
plt.legend()
plt.show()
```

```python
# Testing Set Performance with ADASYN
plt.plot(hist.history['loss'],label='train')
plt.plot(hist.history['val_loss'],label='test')
plt.legend()
plt.show()
```

```python
# Testing Set Performance without smote
plt.plot(hist.history['accuracy'],label='train')
plt.plot(hist.history['val_accuracy'],label='test')
plt.title('Model accuracy during train and test')
plt.xlabel('epochs')
plt.ylabel('accuracy')
```

```python
# Testing Set Performance with ADASYN
plt.plot(hist.history['accuracy'],label='train')
plt.plot(hist.history['val_accuracy'],label='test')
plt.title('Model accuracy during train and test')
plt.xlabel('epochs')
plt.ylabel('accuracy')
```
